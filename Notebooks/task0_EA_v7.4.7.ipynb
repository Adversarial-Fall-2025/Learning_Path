### 1ï¸. Create a Virtual Environment

In the root directory of this repository, open your terminal and run:

- **Windows**
>  ```bash
>  python -m venv .venv
>  ```
- **Linux / macOS**
>  ```bash
>  python3 -m venv .venv
>  ```

Activate the environment:
- **Windows**
>  ```bash
>  .venv\Scripts\activate
>  ```
- **Linux / macOS**
>  ```bash
>  source .venv/bin/activate
>  ```

---

### 2ï¸. Install Dependencies

Once the environment is active, install all required packages:

>```bash
>pip install -r requirements.txt
>```

This will include Jupyter and all necessary notebook dependencies.

---

### 3. Check Jupyter and Python Versions

After dependencies are installed, confirm your setup by running the following commands inside a notebook cell or terminal:

>```bash
>!jupyter --version
>!python --version
>```

Use the **Jupyter version number** (use <u>'notebook'</u>, not Pythonâ€™s) in the notebook filename.  
Each notebook must also include the **entire output** of these commands in one of its Markdown headings or cells for documentation.  
After recording this information, the version check cell can be safely removed.
